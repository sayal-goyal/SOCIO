# Data Preprocessing

```python
column_name = 'category'
values_to_remove = ['EDUCATION', 'CULTURE & ARTS', 'LATINO VOICES', 'COLLEGE', 'ARTS & CULTURE', 'U.S. NEWS', 'GOOD NEWS', 'FIFTY', 'ENVIRONMENT', 'TECH', 'ARTS', 'MONEY', 'TASTE', 'SCIENCE', 'STYLE', 'BLACK VOICES', 'HOME & LIVING', 'PARENTS', 'THE WORLDPOST', 'WEDDINGS', 'WOMEN', 'CRIME', 'IMPACT', 'DIVORCE', 'WORLD NEWS', 'MEDIA', 'WEIRD NEWS', 'GREEN', 'WORLDPOST', 'RELIGION', 'QUEER VOICES', 'PARENTING', 'COMEDY', 'HEALTHY LIVING']
mask = ~df[column_name].isin(values_to_remove)
print(mask)
df = df[mask]
df.drop(columns=['authors','date'], inplace=True)
print("Shape of DataFrame after removing rows:", df.shape)

df.to_json('Modified_News_Category_Dataset.json', orient='records', lines=True)
```

# Data Cleaning

```python
def remove_tags(text):
  remove = re.compile(r'')
  return re.sub(remove, '', text)
df['headline'] = df['headline'].apply(remove_tags)
```

```python
def special_char(text):
  reviews = ''
  for x in text:
    if x.isalnum():
      reviews = reviews + x
    else:
      reviews = reviews + ' '
  return reviews
df['headline'] = df['headline'].apply(special_char)
```

```python
def convert_lower(text):
   return text.lower()
df['headline'] = df['headline'].apply(convert_lower)
df['headline'][1]
```

```python
n1 = LabelEncoder()
df['category'] = n1.fit_transform(df['category'])
df.head(20)
```

```python
X = df['headline']  +  " " + df['short_description']
Y = df['category']
```

```python
count_vect = CountVectorizer()
X_counts = count_vect.fit_transform(X)
```

# Test Train Split

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_counts)
X_train, X_test, Y_train, Y_test = train_test_split(X_train_tfidf, Y, test_size=0.3, random_state=42)
print(X_train)
print(Y_train)
```

# Model Training

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV
start_time = time.time()

mnb = MultinomialNB()
param_grid = {'alpha': [0.1, 0.5, 1.0, 1.5, 2.0]}
grid_search = GridSearchCV(mnb, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, Y_train)
best_params = grid_search.best_params_
best_estimator = grid_search.best_estimator_
predictions = best_estimator.predict(X_test)
accuracy1 = accuracy_score(Y_test, predictions)
print("Accuracy of the Multinomial Naive Bayes model:", accuracy1)

end_time = time.time()
execution_time = end_time - start_time
print('Execution Time:', execution_time, 'seconds')
```

```python
cm  = confusion_matrix(Y_test,predictions)
cr = classification_report(Y_test,predictions)
print(cm)
print("--------------------------------------------------------------------")
print("--------------------------------------------------------------------")
print(cr)
```

```python
from sklearn.linear_model import LogisticRegression
start_time = time.time()

logistic_Regression = LogisticRegression(max_iter=1000, solver='saga')
logistic_Regression.fit(X_train,Y_train)
predictions = logistic_Regression.predict(X_test)
accuracy2 = accuracy_score(Y_test, predictions)
print("Accuracy of the Logistic Regression model:", accuracy2)

end_time = time.time()
execution_time = end_time - start_time
print('Execution Time:', execution_time, 'seconds')
```

```python
from sklearn.neural_network import MLPClassifier
start_time = time.time()

clf_neural = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(10,), random_state=1, max_iter=1000)
clf_neural.fit(X_train, Y_train)
predictions = clf_neural.predict(X_test)
accuracy3 = accuracy_score(Y_test, predictions)
print("Accuracy of the softmax model:", accuracy3)

end_time = time.time()
execution_time = end_time - start_time
print('Execution Time:', execution_time, 'seconds')
```

```python
from sklearn import svm
start_time = time.time()

clf_svm = svm.LinearSVC(dual=False)
clf_svm.fit(X_train, Y_train)
predictions = clf_neural.predict(X_test)
accuracy4 = accuracy_score(Y_test, predictions)
print("Accuracy of the SVM model:", accuracy4)

end_time = time.time()
execution_time = end_time - start_time
print('Execution Time:', execution_time, 'seconds')
```

```python
from sklearn.ensemble import RandomForestClassifier
start_time = time.time()

random_forest = RandomForestClassifier(n_jobs=-1)
random_forest.fit(X_train, Y_train)
predictions = random_forest.predict(X_test)
accuracy5 = accuracy_score(Y_test, predictions)
print("Accuracy of the Random Forest model:", accuracy5)

end_time = time.time()
execution_time = end_time - start_time
print('Execution Time:', execution_time, 'seconds')
```

```python
from sklearn.tree import DecisionTreeClassifier
start_time = time.time()

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)
predictions = decision_tree.predict(X_test)
accuracy6 = accuracy_score(Y_test, predictions)
print("Accuracy of the Decision Tree model:", accuracy6)

end_time = time.time()
execution_time = end_time - start_time
print('Execution Time:', execution_time, 'seconds')
```
